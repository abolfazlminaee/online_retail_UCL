```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_excel('Online Retail.xlsx')
```

```python
data
```

```python
missing_values = data.isna().sum()
print("Missing Values:")
print(missing_values)
```

```python
data.describe()
```

```python
data['Description'] = data['Description'].str.lower()
data.head()
```

```python
data['InvoiceDate'] = pd.to_datetime(data['InvoiceDate'], format='%m/%d/%Y %H:%M')
data['Revenue'] = data['Quantity']*data['UnitPrice']
data['InvoiceMonth'] = data['InvoiceDate'].apply(lambda x: x.strftime('%B'))
data['Day of week'] = data['InvoiceDate'].dt.day_name()
data
```

cancelled & discount

```python
cancellld = data[
    (data['InvoiceNo'].str.startswith('C')) &
    (data['StockCode'] != 'D') &
    (~data['Description'].str.contains('discount', case=False, na=False))
]
cancellld
```

```python
discount = data[
    (data['StockCode'] == 'D') &
    (data['Description'].str.contains('discount', case=False, na=False))
]
discount
```

```python
cancellld.to_csv('data/cancellld.csv', index=False)
discount.to_csv('data/discount.csv', index=False)
data.to_csv('data/data.csv', index=False)
```
